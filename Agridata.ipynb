```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
top_rice = (
    df.groupby("STATE_NAME")["RICE_PRODUCTION_1000_TONS"]
    .sum()
    .sort_values(ascending=False)
    .head(7)
)

top_rice.plot(kind='bar')

plt.title("Top 7 Rice Producing States")
plt.xlabel("State")
plt.ylabel("Rice Production")

plt.xticks(rotation=45)
plt.show()
```

```python
top_wheat = (
    df.groupby("STATE_NAME")["WHEAT_PRODUCTION_1000_TONS"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_wheat.plot(kind='bar')

plt.title("Top 5 Wheat Producing States")
plt.xlabel("State")
plt.ylabel("Wheat Production")

plt.xticks(rotation=45)
plt.show()
```

```python
top_wheat.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Wheat Production Share")
plt.ylabel("")

plt.show()
```

```python
sugar = (
    df.groupby("YEAR")["SUGARCANE_PRODUCTION_1000_TONS"]
    .sum()
)

sugar.plot(kind='line')

plt.title("Sugarcane Production Over Years")
plt.xlabel("Year")
plt.ylabel("Production")

plt.show()
```

```python
rice = (
    df.groupby("YEAR")["RICE_PRODUCTION_1000_TONS"]
    .sum()
)

wheat = (
    df.groupby("YEAR")["WHEAT_PRODUCTION_1000_TONS"]
    .sum()
)

plt.plot(rice, label="Rice")
plt.plot(wheat, label="Wheat")

plt.title("Rice vs Wheat Production")
plt.xlabel("Year")
plt.ylabel("Production")

plt.legend()
plt.show()
```
